Lee notices that the tracker’s battery dies every day at 7 pm. Lee discovers that on an average, he walks 2000 steps every day after 7 pm. Perform an appropriate operation on your array to add 2000 steps to all the observations.

```python
import numpy as np
import pandas as pd

# Create the array
data = np.array([
    [1, 6012],
    [2, 7079],
    [3, 6886],
    [4, 7230],
    [5, 4598],
    [6, 5564],
    [7, 6971],
    [8, 7763],
    [9, 8032],
    [10, 9569]
])

# Add 2000 steps after 7 pm
data[:, 1][data[:, 0] > 7] += 2000

# Filter steps walked if more than 9000
steps_walked_more_than_9000 = data[data[:, 1] > 9000][:, 1]

# Sort steps walked
sorted_steps_walked = np.sort(data[:, 1])

# Print steps walked if more than 9000
print("Steps walked if more than 9000:")
print(steps_walked_more_than_9000)

# Print steps walked in sorted order
print("\nSteps walked in sorted order:")
print(sorted_steps_walked)

# Create a pandas DataFrame
df = pd.DataFrame(data, columns=['Day number', 'Steps walked'])

# Sort steps walked using pandas
sorted_steps_walked_pd = df['Steps walked'].sort_values().values

# Print steps walked in sorted order using pandas
print("\nSteps walked in sorted order using pandas:")
print(sorted_steps_walked_pd)
```

Lee notices that the tracker’s battery dies every day at 7 pm. Lee discovers that on an average, he walks 2000 steps every day after 7 pm. Perform an appropriate operation on your array to add 2000 steps to all the observations.

```python
import numpy as np

# Given data
data = np.array([
    [1, 6012],
    [2, 7079],
    [3, 6886],
    [4, 7230],
    [5, 4598],
    [6, 5564],
    [7, 6971],
    [8, 7763],
    [9, 8032],
    [10, 9569]
])

# Filter the observations after 7 pm and add 2000 steps
data[data[:, 0] > 7, 1] += 2000

# Print the updated array
print(data)
```

In this code, the array data represents the given observations. The operation data[:, 0] > 7 checks for the day number greater than 7, indicating the observations after 7 pm. The resulting Boolean array is used as an index to select the corresponding rows in the second column (data[:, 1]) and add 2000 steps to them.

Pandas also provides us options to visualize the data. Here are some of the examples:

Syntax:

       df.plot(X, y, marker, kind)

       X = value on X axis

       y = value on y axis

       marker = shape in case of specific plots like a scatter plot

       kind = type of plot

Import the data into Python environment as a Pandas DataFrame.
Check for missing values, if any and drop the corresponding rows.
Find the district that gets the highest annual rainfall.
Display the top 5 states that get the highest annual rainfall.
Drop the columns 'Jan-Feb', 'Mar-May', 'Jun-Sep', 'Oct-Dec'.
Display the state-wise mean rainfall for all the months using a pivot table.
Display the count of districts in each state.
For each state, display the district that gets the highest rainfall in May. Also display the recorded rainfall.

```python
import pandas as pd 
data = pd.read_csv('/content/sample_data/rainfall1627650742214.csv')
data.head()
```

```python
print(data.isnull().sum())
```

```python
# Drop rows with missing values
data.dropna(inplace=True)
```

```python
# Find the district with the highest annual rainfall
district_highest_rainfall = data.loc[data['ANNUAL'].idxmax(), 'DISTRICT']
print(district_highest_rainfall)
```

```python
# Display the top 5 states with the highest annual rainfall
top_5_states = data.groupby('STATE_UT_NAME')['ANNUAL'].sum().nlargest(5)
print(top_5_states)
```

```python
# Drop unnecessary columns
data.drop(['Jan-Feb', 'Mar-May', 'Jun-Sep', 'Oct-Dec'], axis=1, inplace=True)
print(data)
```

```python

# State-wise mean rainfall for all the months using a pivot table
state_wise_mean_rainfall = data.pivot_table(index='STATE_UT_NAME', aggfunc='mean')
print(state_wise_mean_rainfall)
```

```python
# Count of districts in each state
district_count_by_state = data['STATE_UT_NAME'].value_counts()
print(district_count_by_state)
```

```python
# District with the highest rainfall in May for each state
district_highest_may_rainfall = data.loc[data.groupby('STATE_UT_NAME')['MAY'].idxmax(), ['STATE_UT_NAME', 'DISTRICT', 'MAY']]
print(district_highest_may_rainfall)
                                         
```
